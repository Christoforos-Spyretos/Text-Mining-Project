```python
import pandas as pd

data = [['MNB CountVectorizer', 94], ['MNB TfidfVectorizer', 94], ['LinearSVM CountVectorizer', 87], 
['LinearSVM TfidfVectorizer', 92], ['TextCNN', 77], ['BiLSTM', 90], ['GRU', 90]]
 
df = pd.DataFrame(data, columns=['Models', 'Accuracies'])
 
df
 
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10,5))
plt.bar(df['Models'], df['Accuracies'], width=0.5)
plt.grid(axis='x')
plt.xticks(rotation=50)
plt.ylim(0,109)
plt.ylabel('Accuracies')
# add values with percentage on top of bars 
for index, value in enumerate(df['Accuracies']):
    plt.text(index, value, str(value) + '%', ha='center')
plt.show()
```
